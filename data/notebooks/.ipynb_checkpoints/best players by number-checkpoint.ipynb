```python
def getBestPlayersByNumber(number, n, gender='male'):  
    number = str(number) 
    n = str(n)
    
    endpoint_url = "https://query.wikidata.org/sparql" 
    
    # Q6581072 - female 
    # Q6581097 - male  
    
    if gender == 'male': 
        sex = 'Q6581097' 
    
    if gender == 'female': 
        sex = 'Q6581072'
    
    
    query = f"""SELECT DISTINCT ?SoccerPlayer ?SoccerPlayerLabel ?number ?sitelinks
    WHERE {{
      ?SoccerPlayer wdt:P106 wd:Q937857; 
                    wdt:P21 wd:{sex};  
                    wdt:P1618 ?number.

      FILTER(?number = '{number}')


      ?SoccerPlayer wikibase:sitelinks ?sitelinks .



      SERVICE wikibase:label {{ bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }}

    }}
    ORDER BY DESC (?sitelinks)
    LIMIT {n}"""
    

    def get_results(endpoint_url, query):
        user_agent = "WDQS-example Python/%s.%s" % (sys.version_info[0], sys.version_info[1])
        # TODO adjust user agent; see https://w.wiki/CX6
        sparql = SPARQLWrapper(endpoint_url, agent=user_agent)
        sparql.setQuery(query)
        sparql.setReturnFormat(JSON)
        return sparql.query().convert()


    results = get_results(endpoint_url, query)  
    
    
    name = [] 
    number = []  
    links = []
    for res in results['results']['bindings']: 
        name.append(res['SoccerPlayerLabel']['value']) 
        number.append(res['number']['value'])   
        links.append(res['sitelinks']['value']) 
        
        
    return pd.DataFrame({
        'Name': name, 
        'Number': number, 
        'Sitelinks': links
    })
```

```python
from math import sin, cos, sqrt, atan2, radians

# approximate radius of earth in km
R = 6373.0

lat1 = radians(52.2296756)
lon1 = radians(21.0122287)
lat2 = radians(52.406374)
lon2 = radians(16.9251681)

dlon = lon2 - lon1
dlat = lat2 - lat1

a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
c = 2 * atan2(sqrt(a), sqrt(1 - a))

distance = R * c

print("Result:", distance)
print("Should be:", 278.546, "km")
```

```python
def getPopPlayersByNationality(Q, n, gender='male', person='player'):
    n = str(n)

    endpoint_url = "https://query.wikidata.org/sparql"

    if gender == 'male':
        sex = 'Q6581097'

    if gender == 'female':
        sex = 'Q6581072'

    if person == 'player':
        p = 'Q937857'

    if person == 'coach':
        p = 'Q628099'

    query = f"""SELECT DISTINCT ?SoccerPlayer ?SoccerPlayerLabel ?soccerNationality ?soccerNationalityLabel ?image
    WHERE {{
      ?SoccerPlayer wdt:P106 wd:{p}; 
                    wdt:P21 wd:{sex};   
                    wdt:P1532 ?soccerNationality.

      FILTER(?soccerNationality = wd:{Q}) 

      OPTIONAL{{?SoccerPlayer wdt:P18 ?image.}}

      ?SoccerPlayer wikibase:sitelinks ?sitelinks .


      SERVICE wikibase:label {{ bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }}

    }} 
    ORDER BY DESC (?sitelinks)
    LIMIT {n}"""

    def get_results(endpoint_url, query):
        user_agent = "WDQS-example Python/%s.%s" % (sys.version_info[0], sys.version_info[1])
        # TODO adjust user agent; see https://w.wiki/CX6
        sparql = SPARQLWrapper(endpoint_url, agent=user_agent)
        sparql.setQuery(query)
        sparql.setReturnFormat(JSON)
        return sparql.query().convert()

    results = get_results(endpoint_url, query)

    name = []
    img = []
    for res in results['results']['bindings']:
        name.append(res['SoccerPlayerLabel']['value'])
        try:
            img.append(res['image']['value'])
        except:
            img.append(np.nan)

    # pop_players = pd.DataFrame({
    #     'name': name,
    #     'image': [html.Center(html.Img(src=x,
    #                                    style={'width': '30%', 'height':'20px'})) for x in img]
    # })

    pop_players = pd.DataFrame({
        'name': name
    })

    return pop_players


```

```python
def generalMostPopular(n, gender='male'):
    n = str(n)

    endpoint_url = "https://query.wikidata.org/sparql"

    if gender == 'male':
        sex = 'Q6581097'

    if gender == 'female':
        sex = 'Q6581072'

    query = f"""SELECT DISTINCT ?SoccerPlayer ?SoccerPlayerLabel ?number ?sitelinks ?image
    WHERE {{
      ?SoccerPlayer wdt:P106 wd:Q937857; 
                    wdt:P21 wd:{sex};   

      OPTIONAL{{?SoccerPlayer wdt:P18 ?image.}}

      ?SoccerPlayer wikibase:sitelinks ?sitelinks .


      SERVICE wikibase:label {{ bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }}

    }}
    ORDER BY DESC (?sitelinks)
    LIMIT {n}"""

    def get_results(endpoint_url, query):
        user_agent = "WDQS-example Python/%s.%s" % (sys.version_info[0], sys.version_info[1])
        # TODO adjust user agent; see https://w.wiki/CX6
        sparql = SPARQLWrapper(endpoint_url, agent=user_agent)
        sparql.setQuery(query)
        sparql.setReturnFormat(JSON)
        return sparql.query().convert()

    results = get_results(endpoint_url, query)

    name = []
    links = []
    img = []
    for res in results['results']['bindings']:
        name.append(res['SoccerPlayerLabel']['value'])
        links.append(res['sitelinks']['value'])
        try:
            img.append(res['image']['value'])
        except:
            img.append(np.nan)

    return pd.DataFrame({
        'Name': name,
        'Sitelinks': links
    })
```
